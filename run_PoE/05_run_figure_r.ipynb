### ****Plot the heatmap****

```r
slisa_biclust = read.csv('~/Spatial_project/results/03_hubs/PoE_alltumors_bicluster.csv',row.names = 1)

#slisa_biclust=read.csv('~/Spatial_project/results/03_hubs/PoE_alltumors_factor',row.names = 1)
#slisa_biclust_s = slisa_biclust[names(which(apply(abs(slisa_biclust), 1, max)>1)),]
slisa_biclust_s = slisa_biclust
row_cl = read.csv('~/Spatial_project/results/03_hubs/PoE_alltumors_bicluster_res_row_orders_individual.csv')
#head(row_cl)
row.names(row_cl) = row_cl$id
#table(row_cl$cluster)

col_cl = read.csv('~/Spatial_project/results/03_hubs/PoE_alltumors_bicluster_res_col_orders_individual.csv')
#head(col_cl)
row.names(col_cl) = col_cl$id
#table(col_cl$cluster)
new_col = c()


bn_dict= data.frame(col_cl$id,
                     c(paste0('X',seq(0,25))))


colnames(bn_dict) = c('cell_type','bn')

row_cl$cluster = as.factor(row_cl$cluster)

cts_df = read.csv('~/Spatial_project/results/03_hubs_0106/all_ct_tumors.csv',row.names = 1)
source_df = read.csv('~/Spatial_project/results/03_hubs_0106/all_source_tumors.csv',row.names = 1)


#cts_df = data.frame(cts_df[names(which(apply(abs(slisa_biclust), 1, max)>1)),,])

row.names(cts_df) = row.names(row_cl)
row.names(source_df) = row.names(row_cl)

colnames(cts_df) = 'x'
colnames(source_df) = 'source'

cts_df$x = as.factor(cts_df$x)
source_df$source = as.factor(source_df$source)

col1 = structure(c('steelblue1',
                   'slateblue1'
                   ),
                 names=c('metaplastic',
                         'TNBC'
                         ))


col2 = structure(c(
                   'slateblue',
                    #'darkorange',
                   'slateblue',#'darkorange',
                   'darkolivegreen1',
                   'darkolivegreen1',#'darksalmon',
                   'deepskyblue',
                   'deepskyblue',#'orangered',
                   'coral4',
                   'coral4', #'coral1',
                   'darkgoldenrod1',
                   'darkgoldenrod1',#'mediumorchid1',
                   'olivedrab',
                   'olivedrab',#'mediumorchid',
                   'orangered',
                   'orangered'#'seagreen2'
),
                 names=c('MBC_1A',
                         'MBC_1B',
                         'MBC_2A',
                         'MBC_2B',
                         'MBC_3A',
                         'MBC_3B',
                         'MBC_4A',
                         'MBC_4B',
                         'CID4465',
                         'CID444971',
                         '1142243F',
                         '1160920F' 
                        )
                )

h1 = Heatmap(data.matrix(slisa_biclust_s),
             show_row_names = FALSE,
             row_dend_reorder = FALSE,
             column_dend_reorder = FALSE,
             row_order = row.names(slisa_biclust_s),
             column_order = colnames(slisa_biclust_s),
             #col = colorRamp2(c(-2, 0,2), c("blue", "white", "red")),
             name = 'bn')

h2 = Heatmap(source_df[row.names(slisa_biclust_s),'source'],show_row_names = FALSE,row_dend_reorder = FALSE,column_dend_reorder = FALSE,name='sample_id',col=col2)#,col=colors
h3 = Heatmap(cts_df[row.names(slisa_biclust_s),'x'],show_row_names = FALSE,row_dend_reorder = FALSE,column_dend_reorder = FALSE,name='cancer_type',col=col1)#,col=colors
h4 = Heatmap(row_cl[row.names(slisa_biclust_s),'cluster'],show_row_names = FALSE,row_dend_reorder = FALSE,column_dend_reorder = FALSE,
             col=c('blue',
                   'orange',
                   'darkolivegreen1',
                   'red',
                   'brown',
                   'pink',
                   'darkolivegreen4',
                   'cyan',
                   'coral1',
                   'slateblue',
                   'black'
                   #'mediumorchid1'
                   
                  ),name='hub')#,'green''darkgoldenrod1',
draw(h1+h2+h3+h4)
```
